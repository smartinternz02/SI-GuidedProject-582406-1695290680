## **Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)**

```python
import pandas as pd
import numpy as np
import random
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
data={
    'Matches':np.random.randint(0,200,size=4),
    'Runs Scored':np.random.randint(100,8000,size=4),
    'Average':np.random.uniform(0.0,100.0,4),
    'Central Contract':[random.choice(list(alphabet)) for _ in range(4)]
}
df=pd.DataFrame(data)
df

```

# **Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4**

```python
df.rename(columns={'Matches':'Random value 1','Runs Scored':'Random value 2','Average':'Random value 3','Central Contract':'Random value 4'},inplace=True)
df
```
